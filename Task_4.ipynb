# Условие 3: 3 задача
## Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
plt.figure(figsize=(9, 9))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
plt.ylabel('Площадь')
plt.xlabel('Цена');
```

Имеется прямая зависимоть цены от жилой площади - чем больше площадь, тем выше цена

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
plt.ylabel('Категория постройки')
plt.xlabel('Цена');
```

Чем выше категория постройки, тем выше цена (элитные дома - самые дорогие)

```python
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), showfliers = False);
plt.xlabel('Цена')
plt.ylabel('Состояние (оценка)')
plt.title('Зависимость цены от вида на набережную');
```

При равных прочих, наболее дорогими являются дома с видом на набережную

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), showfliers=False)
plt.xlabel('price')
plt.ylabel('view')
plt.title('Зависимость цены от внешнего вида дома')
```

Вид дома, с оценкой "4", имеет максимальную стоимость 

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), showfliers=False);
plt.xlabel('Цена')
plt.ylabel('Этажи')
plt.title('Зависимость цены от кол-ва этажей');
```

Наибольшую стоимость имеют дома с количеством этажей 2,5
